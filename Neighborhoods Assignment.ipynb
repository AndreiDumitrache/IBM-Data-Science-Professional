```python
import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
df
```

```python
df1 = df[~df["Borough"].str.contains('Not assigned')]
df1
```

```python
df1 = df1.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df1
```

```python
df1.loc[df1['Neighbourhood']=="Not assigned",'Neighbourhood']=df1.loc[df1['Neighbourhood']=="Not assigned",'Borough']
df1
```

```python
df1.shape
```

I have turned the table from the Wikipedia page into a pandas Dataframe, and then proceeded to remove the cells in which the Borough was not assigned. After that I have combined the Neighbourhoods where the Postal code was the same. Then, for every Not assigned Neighbourhood I have changed the Not assigned cell with the corresponding Borough cell.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
        venues_list=[]
        for name, lat, lng in zip(names, latitudes, longitudes):
            print(name)
            
            url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
            results = requests.get(url).json()['response']['groups'][0]['items']
            
            venues_list.append([(
            name,
            lat,
            lng,
            v['venue']['name'],
            v['venue']['location']['lat'],
            v['venue']['location']['lng'],
            v['venue']['categories'][0]['name']) for v in results])
            
        nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
        return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=df2['Neighbourhood'],
                              latitudes = df2['Latitude'],
                              longitudes = df2['Longitude'])
```

The 'toronto_venues' dataframe contains all venues found through the Foursquare API in Toronto (all neighbourhoods and all Boroughs from the df2 dataframe).

```python
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix = '', prefix_sep = '')
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood']
toronto_onehot.head()
```

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean()
toronto_grouped
```

```python
toronto_grouped.shape
```

```python
kclusters = 5
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped)
kmeans.labels_[0:20]
```

```python
map_clusters = folium.Map(location=[43.807448, -79.199056], zoom_start=11)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []

folium.CircleMarker(
        [43.806686, -79.194353],
        radius=5,
        color=rainbow,
        fill=True,
        fill_color=rainbow,
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
